```python
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report

# Load penguin data
penguin_data = pd.read_csv('/content/penguins_lter.csv')

# Replace NaN values with the mean or use your preferred method for handling missing values
penguin_data = penguin_data.fillna(penguin_data.mean())

# Select relevant columns
selected_columns = ['Culmen Length (mm)', 'Culmen Depth (mm)', 'Sex']
penguin_subset = penguin_data[selected_columns]

# Calculate correlation matrix
correlation_matrix = penguin_subset.corr()

# Visualize correlation using a heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")

# Split the data into features (X) and target (y)
X = penguin_subset[['Culmen Length (mm)', 'Culmen Depth (mm)']]
y = penguin_subset['Sex']

# Encode the target variable 'Sex'
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42)

# Build a Random Forest Classifier model
model = RandomForestClassifier(random_state=42)
model.fit(X_train, y_train)

# Make predictions on the test set
y_pred = model.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy:.2f}')
print('Classification Report:')
print(classification_report(y_test, y_pred))
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report

# Load penguin data
penguin_data = pd.read_csv('/content/penguins_lter.csv')

# Replace NaN values with the mean or use your preferred method for handling missing values
penguin_data = penguin_data.fillna(penguin_data.mean())

# Select relevant columns
selected_columns = ['Culmen Length (mm)', 'Culmen Depth (mm)', 'Sex']
penguin_subset = penguin_data[selected_columns]

# Split the data into features (X) and target (y)
X = penguin_subset[['Culmen Length (mm)', 'Culmen Depth (mm)']]
y = penguin_subset['Sex']

# Encode the target variable 'Sex'
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42)

# Feature Scaling
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Build a k-Nearest Neighbors (KNN) model
knn_model = KNeighborsClassifier(n_neighbors=3)  # You can adjust the number of neighbors (n_neighbors) as needed
knn_model.fit(X_train_scaled, y_train)

# Make predictions on the test set
y_pred = knn_model.predict(X_test_scaled)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy:.2f}')
print('Classification Report:')
print(classification_report(y_test, y_pred))

# Visualize decision regions
h = .02  # Step size in the mesh
x_min, x_max = X_train_scaled[:, 0].min() - 1, X_train_scaled[:, 0].max() + 1
y_min, y_max = X_train_scaled[:, 1].min() - 1, X_train_scaled[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

fig, ax = plt.subplots()
plt.tight_layout(h_pad=0.5, w_pad=0.5, pad=2.5)

Z = knn_model.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

ax.contourf(xx, yy, Z, cmap=plt.cm.coolwarm, alpha=0.8)
ax.scatter(X_train_scaled[:, 0], X_train_scaled[:, 1], c=y_train, edgecolors='k', cmap=plt.cm.coolwarm, s=20, marker='o', label='Train Set')
ax.scatter(X_test_scaled[:, 0], X_test_scaled[:, 1], c=y_test, edgecolors='k', cmap=plt.cm.coolwarm, s=40, marker='x', label='Test Set')

ax.set_xlabel('Culmen Length (mm)')
ax.set_ylabel('Culmen Depth (mm)')
ax.set_title('Decision Boundary of KNN')

plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report

# 펭귄 데이터 불러오기
penguin_data = pd.read_csv('/content/penguins_lter.csv')

# NaN 값들을 평균값으로 대체하거나 누락된 값 처리를 위한 선호하는 방법 사용
penguin_data = penguin_data.fillna(penguin_data.mean())

# 관련 있는 열 선택
selected_columns = ['Culmen Length (mm)', 'Culmen Depth (mm)', 'Sex']
penguin_subset = penguin_data[selected_columns]
# Calculate correlation matrix
correlation_matrix = penguin_subset.corr()

# Visualize correlation using a heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")

# 데이터를 특성 (X)과 타겟 (y)으로 나누기
X = penguin_subset[['Culmen Length (mm)', 'Culmen Depth (mm)']]
y = penguin_subset['Sex']

# 타겟 변수 '성별'을 인코딩
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

# 데이터를 훈련 및 테스트 세트로 나누기
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42)

# 랜덤 포레스트 분류기 모델 만들기
rf_model = RandomForestClassifier(random_state=42)
rf_model.fit(X_train, y_train)

# 테스트 세트에 대해 예측하기
y_pred = rf_model.predict(X_test)

# 모델 평가
accuracy = accuracy_score(y_test, y_pred)
print(f'정확도: {accuracy:.2f}')
print('분류 보고서:')
print(classification_report(y_test, y_pred))

# 결정 영역 시각화
h = .02  # 메쉬의 스텝 크기
x_min, x_max = X_train.iloc[:, 0].min() - 1, X_train.iloc[:, 0].max() + 1
y_min, y_max = X_train.iloc[:, 1].min() - 1, X_train.iloc[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

fig, ax = plt.subplots()
plt.tight_layout(h_pad=0.5, w_pad=0.5, pad=2.5)

Z = rf_model.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# 결정 경계를 나타내는 등고선 플롯
ax.contourf(xx, yy, Z, cmap=plt.cm.coolwarm, alpha=0.8)

# 훈련 세트의 산점도 플롯
ax.scatter(X_train.iloc[:, 0], X_train.iloc[:, 1], c=y_train, edgecolors='k', cmap=plt.cm.coolwarm, s=20, marker='o', label='훈련 세트')

# 테스트 세트의 산점도 플롯
ax.scatter(X_test.iloc[:, 0], X_test.iloc[:, 1], c=y_test, edgecolors='k', cmap=plt.cm.coolwarm, s=40, marker='x', label='테스트 세트')

# x 및 y 축 레이블 및 제목 설정
ax.set_xlabel('Culmen Length (mm)')
ax.set_ylabel('Culmen Depth (mm)')
ax.set_title('Decision Boundary of Random Forest')
# 플롯 보이기
plt.show()
```
