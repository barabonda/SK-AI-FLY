```python
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
import newspaper as news
from konlpy.tag import Hannanum

# 불용어 추가 또는 수정
custom_stopwords = set(STOPWORDS)
custom_stopwords.update(['있다', '등', '것으로','외벽', '고', '각', '말까지', '이같이', '으로', '수', '만큼', '다른', '등이다', '의', '당', 'CC', '약', '중이다','합니다', '하는', '할', '하고', '한다',
             '순간','최대','이날','기준','지난달','그리고', '입니다', '그', '등', '이런','이', '및','제', '더'])

link = 'https://v.daum.net/v/20231225123129182'
article = news.Article(link, language='ko')
article.download()
article.parse()
article_text = article.text

# 형태소 분석기를 사용하여 명사 추출
hannanum = Hannanum()
nouns = hannanum.nouns(article_text)

# Load the mask image
mask_image = np.array(Image.open("/content/피카츄.jpg"))
# Generate WordCloud with mask and custom stopwords
wordcloud = WordCloud(
    max_words=100,
    font_path='/usr/share/fonts/truetype/nanum/NanumGothicBold.ttf',
    stopwords=custom_stopwords,
    mask=mask_image,
    background_color='black'
).generate(" ".join(nouns))
# Display the WordCloud
plt.figure(figsize=(5, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
