# URL Document Loader

LangChain은 웹에 기록된 글도 텍스트 형식으로 가져와서 LLM에 활용할 수 있습니다.



```python
from langchain.document_loaders import WebBaseLoader

loader = WebBaseLoader('https://www.nocutnews.co.kr/news/6087302')
data = loader.load()

print(data[0].page_content)
```
